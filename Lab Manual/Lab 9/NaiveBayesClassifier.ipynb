# Task 1

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
x = iris.data
y = iris.target

df = pd.DataFrame(x, columns=iris.feature_names)
df['target'] = y
df
```

# Task 2

```python
print("Missing Values Check")
df.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
print("Data Preprocessing Done!\nScaling Applied.")
```

# Task 3

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Training set size: {x_train.shape[0]} samples")
print(f"Testing set size: {x_test.shape[0]} samples")
```

# Task 4-5

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(x_train, y_train)

print("Gaussian Naïve Bayes Model Trained.")

y_pred = gnb.predict(x_test)
print("Predictions Made on Test Set.")
```

# Task 6

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(y_test, y_pred)
print("Model Performance")
print(f"Accuracy: {accuracy:.4f}\n")

print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))
```

# Task 7

```python
import numpy as np

new_data_unseen = np.array([[5.1, 3.5, 1.4, 0.2]])
new_data_scaled = scaler.transform(new_data_unseen)

prediction = gnb.predict(new_data_scaled)
prediction_probability = gnb.predict_proba(new_data_scaled)

print("Testing on New Unseen Data")
print(f"New Flower Features: {new_data_unseen[0]}")
print(f"Predicted Class Index: {prediction[0]}")
print(f"Predicted Species: {iris.target_names[prediction[0]]}")

print("\nPrediction Probabilities:")
prob_df = pd.DataFrame(prediction_probability, columns=iris.target_names, index=['Probability'])
prob_df
```
